```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
from google.colab import drive

drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/My Drive/house/data.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
X = df[['sqft_lot','bedrooms','bathrooms']]
Y = df['price']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()

model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
def predict_price(sqft_lot, bedrooms, bathrooms):
    input_data = pd.DataFrame({
        'sqft_lot': [sqft_lot],
        'bedrooms': [bedrooms],
        'bathrooms': [bathrooms]
    })

    predicted_price = model.predict(input_data)

    return predicted_price[0]

sqft_lot = float(input("Enter the square footage (sqft_lot): "))
bedrooms = int(input("Enter the number of bedrooms: "))
bathrooms = float(input("Enter the number of bathrooms: "))

predicted_price = predict_price(sqft_lot, bedrooms, bathrooms)
print(f'The predicted price for a house with {sqft_lot} square feet, {bedrooms} bedrooms, and {bathrooms} bathrooms is ${predicted_price:.2f}')
```
